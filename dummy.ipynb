```python
import ChordDetection
import os
import pandas as pd
import numpy as np
from KeyDetection import key_detection_krumhansl_shmuckler
import os
```

```python
def generate_chord_detection_csv_files(time_list, final_preds_chords):
    start = time_list[0]
    df_list = []
    for i in range(0, len(time_list)):
        if(i == len(time_list) - 2): #penultimo seg
            df_list.append([start, time_list[i + 1], final_preds_chords[i]])
            break
        elif(final_preds_chords[i] != final_preds_chords[i + 1]):
                df_list.append([start, time_list[i], final_preds_chords[i]])
                start = time_list[i + 1]

    for i in range(1, len(df_list)):
        df_list[i][0] = df_list[i-1][1]

    df_list = [tuple(seg) for seg in df_list]

    df = pd.DataFrame(df_list, columns=["start", "end", "chord"])

    return df

def get_list_progression(song_path):

    chord_list = ['A','A#','B','C','C#','D','D#','E','F','F#','G','G#','Am','A#m','Bm','Cm','C#m','Dm','D#m','Em','Fm','F#m','Gm','G#m','N']

    DC = ChordDetection.DetectChords()
    DC.build_cnn_extractor()
    DC.build_crf()

    if os.path.exists(song_path.replace(".mp3", ".wav")):
        os.remove(song_path.replace(".mp3", ".wav"))

    DC.mp3_to_wav(song_path)

    predictions_shape, final_preds = DC.predict_seq()

    start = 0
    end = predictions_shape[1]
    time_list = np.round(np.arange(start//10, end//10, 0.1), 1)

    final_preds_chords = [chord_list[int(chord_idx)] for chord_idx in final_preds]
    final_preds_chords = final_preds_chords[0: len(time_list)]

    return generate_chord_detection_csv_files(time_list, final_preds_chords)
```

```python
dataset = get_list_progression("Set1/001.mp3")
```

```python
folder = "set2"
files = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]

mean_ratings_set1 = pd.read_csv("mean_ratings_set2.csv")

prog_emo = pd.DataFrame(columns=["progression","key","song","valence","energy","tension","anger","fear","happy","sad","tender","TARGET"])

for file in files:
    try: 
        idx = int(file.replace(".mp3", ""))
        data = mean_ratings_set1[mean_ratings_set1["Number"] == idx].to_dict("records")[0]
        data["key"] = key_detection_krumhansl_shmuckler(folder + "/" + file)
        data["progression"] = get_list_progression(folder + "/" + file)["chord"].to_list()

        data_selected = {
            "progression": data["progression"],
            "key": data["key"],
            "song": data["Number"],
            "valence": data["Valence"],
            "energy": data["Energy"],
            "tension": data["Tension"],
            "anger": data["Anger"],
            "fear": data["Fear"],
            "happy":data["Happy"],
            "sad":data["Sad"],
            "tender": data["Tender"],
            "TARGET": data["TARGET"]
        }

        prog_emo = pd.concat([prog_emo, pd.DataFrame([data_selected])])
    except:
        # save the file name in a txt
        with open("error_files.txt", "a") as f:
            f.write(file + "\n")
 
```

```python
prog_emo.to_csv("prog_emo_2.csv", index=False)
```
